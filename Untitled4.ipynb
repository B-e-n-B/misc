```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

url = "https://raw.githubusercontent.com/alexbernal0/QEP/main/CrudeOpen2Open.csv"
obsperfdf = pd.read_csv(url)

# drop the "Date" column
obsperfdf = obsperfdf.drop(['Date'], axis = 1)

# create a new column 'Profit'
obsperfdf['Profit'] = ""

# calculate the profit based on the 'Signal' column
obsperfdf.loc[obsperfdf['Signal'] == 'Buy', 'Profit'] = (obsperfdf['OpenExitPrice'] - obsperfdf['OpenPrice']) * 1000
obsperfdf.loc[obsperfdf['Signal'] == 'Short', 'Profit'] = (obsperfdf['OpenPrice'] - obsperfdf['OpenExitPrice']) * 1000

#using 40k ($6000 margin) to based 15% Margin to equity
obsperfdf['CumPL'] = 40000 + obsperfdf['Profit'].cumsum()

# Convert the 'Profit/Loss' and 'equity' columns to numeric data type
obsperfdf['Profit'] = pd.to_numeric(obsperfdf['Profit'], errors='coerce')
obsperfdf['CumPL'] = pd.to_numeric(obsperfdf['CumPL'], errors='coerce')

# Create a new column called 'returnPerc'
obsperfdf['rtnPerc'] = obsperfdf['CumPL'].pct_change()

# Set the first value of 'returnPerc' to fill NaN
obsperfdf.at[obsperfdf.index[0], 'rtnPerc'] = -0.025

# Calculate the drawdown
cumulative_return = (1 + obsperfdf['rtnPerc']).cumprod()
drawdown = (cumulative_return.cummax() - cumulative_return) / cumulative_return.cummax()

# Add a new column for drawdown
obsperfdf['Drawdown'] = -drawdown

# drop some columns
obsperfdf = obsperfdf.drop(['#','Signal','Profit','OpenPrice','OpenExitPrice','rtnPerc','Drawdown'], axis = 1)

# rename some columns
obsperfdf = obsperfdf.rename(columns={'Exit Date/Time': 'Date'})
obsperfdf = obsperfdf.rename(columns={'CumPL': 'Close'})

# Convert the exit date time column to a pandas datetime object and set it as the index
obsperfdf['Date'] = pd.to_datetime(obsperfdf['Date'])
obsperfdf.set_index('Date', inplace=True)

# Resample the dataframe into daily frequency
obsperfdfn = obsperfdf.resample('D').last()

# Replace NaN values with last recorded price
obsperfdfn = obsperfdfn.fillna(method='ffill')

obsperfdfn
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
import scipy.optimize as opt
from scipy.optimize import minimize

########################################################################################################################################
################THIS SECTION IS FOR PLOTTING THE EFFICIENT FRONTEIR FOR THE MONOLITH ASSET GROUP########################################
########################################################################################################################################

# Bring in data
obsidiandataf = pd.read_csv('obsidiandata')
spydataf = pd.read_csv('spydata')
aggdataf = pd.read_csv('aggdata')

# Define the 'close' rows of each dataframe
obsidiandataf_prices = obsidiandataf['Close']
spydataf_prices = spydataf['Close']
aggdataf_prices = aggdataf['Close']

# Define the data that will be used to make an efficient fronteir
obs_portfolio_data = pd.concat([obsidiandataf_prices,spydataf_prices,aggdataf_prices], axis = 1)
obs_portfolio_data.columns = ['Obsidian Performance','SPY','AGG']

# Calculate portfolio returns
firstreturns = obs_portfolio_data.pct_change().dropna()

# Calculate the mean returns and covariance matrix
firstmu = firstreturns.mean()
firstsigma = firstreturns.cov()

# Define the number of points to plot on the efficient frontier
firstnum_points = 50

# Generate a range of portfolio weights
firstweights = np.linspace(0, 1, firstnum_points)
firstweights = np.append(firstweights, 1) # Add 100% allocation toaggdataf

# Calculate the expected returns and volatilities for each portfolio
firstport_returns = []
firstport_volatilities = []
for firstw in firstweights:
    firstw_third = 1 - firstw    
    port_return = (1 - firstw - firstw_third) * firstmu['SPY'] + firstw * firstmu['AGG'] + firstw_third * firstmu['Obsidian Performance']
    port_volatility = np.sqrt(
        (firstw ** 2) * firstsigma.loc['AGG', 'AGG']
        + ((1 - firstw - firstw_third) ** 2) * firstsigma.loc['SPY', 'SPY']
        + (firstw_third ** 2) * firstsigma.loc['Obsidian Performance', 'Obsidian Performance']
        + 2 * firstw * (1 - firstw - firstw_third) * firstsigma.loc['AGG', 'SPY']
        + 2 * firstw_third * (1 - firstw - firstw_third) * firstsigma.loc['AGG', 'Obsidian Performance']
        + 2 * firstw * firstw_third * firstsigma.loc['SPY', 'Obsidian Performance'])
    firstport_returns.append(port_return)
    firstport_volatilities.append(port_volatility)


# Convert daily returns to annualized returns
firstann_returns = firstmu * 252

# Calculate the annualized expected returns for each portfolio
firstann_port_returns = []
for firstret in firstport_returns:
    firstann_port_return = firstret * 252
    firstann_port_returns.append(firstann_port_return)


########################################################################################################################################
#######################THIS SECTION IS FOR PLOTTING THE EFFICIENT FRONTEIR FOR THE SPY/AGG ASSET GROUP #################################
########################################################################################################################################


# Define the tickers
tickers = ['SPY', 'AGG']

# Retrieve historical stock data
data = yf.download(tickers, start='2016-01-01', end='2021-12-31')['Adj Close']

# Calculate the daily returns
returns = data.pct_change().dropna()

# Calculate the mean returns and covariance matrix
mu = returns.mean()
sigma = returns.cov()

# Define the number of points to plot on the efficient frontier
num_points = 50

# Generate a range of portfolio weights
weights = np.linspace(0, 1, num_points)
weights = np.append(weights, 1) # Add 100% allocation to AGG

# Calculate the expected returns and volatilities for each portfolio
port_returns = []
port_volatilities = []
for w in weights:
    port_return = (1 - w) * mu['SPY'] + w * mu['AGG']  # Calculate the portfolio return
    port_volatility = np.sqrt((w ** 2) * sigma.loc['AGG', 'AGG'] + ((1 - w) ** 2) * sigma.loc['SPY', 'SPY'] + 2 * w * (1 - w) * sigma.loc['AGG', 'SPY'])
    port_returns.append(port_return)
    port_volatilities.append(port_volatility)

# Convert daily returns to annualized returns
ann_returns = mu * 252

# Calculate the annualized expected returns for each portfolio
ann_port_returns = []
for ret in port_returns:
    ann_port_return = ret * 252
    ann_port_returns.append(ann_port_return)

# Find the index of the portfolios with 100% allocation to SPY and AGG
spy_alloc_index = np.where(weights == 0)[0][0]
agg_alloc_index = np.where(weights == 1)[0][0]


########################################################################################################################################
#######################THIS SECTION PLOTS THE PORTFOLIO MARKERS FOR THE MONOLITH ASSET GROUP############################################
########################################################################################################################################


# Bring in data
obsidiandataf = pd.read_csv('obsidiandata')
spydataf = pd.read_csv('spydata')
aggdataf = pd.read_csv('aggdata')

# Define the 'close' rows of each dataframe
obsidiandataf_prices = obsidiandataf['Close']
spydataf_prices = spydataf['Close']
aggdataf_prices = aggdataf['Close']

markerdata = pd.concat([obsidiandataf_prices, spydataf_prices, aggdataf_prices], axis=1)
markerdata.columns = ['Obsidian Performance', 'SPY', 'AGG']

# Calculate returns and covariance matrix
markerreturns = markerdata.pct_change().dropna()
markercov_matrix = markerreturns.cov()

# Function to calculate portfolio markerreturns and volatility
def markercalculate_portfolio_performance(markerweights, markerreturns, markercov_matrix):
    markerweights = np.array(markerweights)  # Convert markerweights to a NumPy array
    markerportfolio_return = np.sum(markerreturns.mean() * markerweights) * 252
    markerportfolio_std_dev = np.sqrt(np.dot(markerweights.T, np.dot(markercov_matrix, markerweights))) * np.sqrt(252)
    return markerportfolio_return, markerportfolio_std_dev

# Function to minimize the negative portfolio Sharpe ratio
def markerminimize_negative_sharpe_ratio(markerweights, markerreturns, markercov_matrix, risk_free_rate):
    markerportfolio_return, markerportfolio_std_dev = markercalculate_portfolio_performance(markerweights, markerreturns, markercov_matrix)
    markersharpe_ratio = (markerportfolio_return - risk_free_rate) / markerportfolio_std_dev
    return -markersharpe_ratio

# Set the risk-free rate
risk_free_rate = 0.02

# Define optimization constraints
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# Define optimization bounds
bounds = tuple((0, 1) for _ in range(len(markerdata.columns)))

# Initial guess for markerweights
markerinitial_weights = [1/len(markerdata.columns)] * len(markerdata.columns)  # Equal allocation

# Optimize portfolio markerweights for maximum Sharpe ratio
markerresult = minimize(markerminimize_negative_sharpe_ratio, markerinitial_weights,
                  args=(markerreturns, markercov_matrix, risk_free_rate),
                  method='SLSQP', bounds=bounds, constraints=constraints)

# Get optimized markerweights
markeroptimized_weights = markerresult.x

# Calculate optimized portfolio performance
markeroptimized_return, markeroptimized_std_dev = markercalculate_portfolio_performance(
    markeroptimized_weights, markerreturns, markercov_matrix)

# Generate random portfolios
markernum_portfolios = 100
markerall_weights = np.zeros((markernum_portfolios, len(markerdata.columns)))
markerret_arr = np.zeros(markernum_portfolios)
markervol_arr = np.zeros(markernum_portfolios)

for i in range(markernum_portfolios):
    markerweights = np.random.random(len(markerdata.columns))
    markerweights /= np.sum(markerweights)
    markerall_weights[i, :] = markerweights
    markerret_arr[i], markervol_arr[i] = markercalculate_portfolio_performance(markerweights, markerreturns, markercov_matrix)

# Calculate performance of market portfolios
market_weights = [0.1, 0.6, 0.3]
market_weights_1 = [0.1, 0.54, 0.36]
market_weights_2 = [0.1, 0.5, 0.4]

# Calculation of portfolio weights
market_return, market_std_dev = markercalculate_portfolio_performance(market_weights, markerreturns, markercov_matrix)
market_return_1, market_std_dev_1 = markercalculate_portfolio_performance(market_weights_1, markerreturns, markercov_matrix)
market_return_2, market_std_dev_2 = markercalculate_portfolio_performance(market_weights_2, markerreturns, markercov_matrix)


########################################################################################################################################
######################THIS SECTION HANDLES PLOTTING FOR ALL OF THE GRAPHICS#############################################################
########################################################################################################################################


#-------MONOLITH EFFICIENT FRONTEIR-------------------

# Plot the efficient frontier
plt.figure(figsize=(8, 6))
plt.plot(firstport_volatilities, firstann_port_returns, 'b-', label='Efficient Frontier')

#-------SPY/AGG EFFICIENT FRONTEIR-------------------

# Plot the efficient frontier
plt.plot(port_volatilities, ann_port_returns, 'b-')

# Add markers to highlight the portfolios 
plt.plot(port_volatilities[spy_alloc_index], ann_port_returns[spy_alloc_index], color = 'steelblue', marker = 'o', alpha=0.9)
plt.plot(port_volatilities[agg_alloc_index], ann_port_returns[agg_alloc_index], color = 'steelblue', marker = 'o', alpha=0.9)

#annotate different porfolios on the fronteir
plt.annotate('100% Stocks', xy=(port_volatilities[spy_alloc_index], ann_port_returns[spy_alloc_index]),
             xytext=(0, 20), textcoords='offset points', ha='left', va='bottom',
             arrowprops=dict(arrowstyle='-', connectionstyle='arc3,rad=-0.2', color='grey'))
plt.annotate('100% Bonds', xy=(port_volatilities[agg_alloc_index], ann_port_returns[agg_alloc_index]),
             xytext=(20, 0), textcoords='offset points', ha='left', va='top',
             arrowprops=dict(arrowstyle='-', connectionstyle='arc3,rad=-0.2', color='grey'))

#-------MONOLITH PORTFOLIO MARKERS-------------------

# Plot efficient frontier
plt.plot(markervol_arr, markerret_arr, 'o', markersize=5, label='Portfolios')

# Plot markers and add annotations to each marker
plt.scatter(markeroptimized_std_dev, markeroptimized_return, marker='D', color='orange', s=100, label='Optimized Portfolio')
plt.annotate('60% Stocks, 30% Bonds, \n10% Monolith', (markeroptimized_std_dev, markeroptimized_return), textcoords="offset points", xytext=(-10,-10), ha='center', 
arrowprops=dict(arrowstyle='-', connectionstyle='arc3,rad=-0.2', color='grey'))

plt.scatter(market_std_dev_1, market_return_1, marker='D', color='orange', s=100, label='Market Portfolio 1')
plt.annotate('54% Stocks, 36% Bonds, \n10% Monolith', (market_std_dev_1, market_return_1), textcoords="offset points", xytext=(-10,-10), ha='center', 
arrowprops=dict(arrowstyle='-', connectionstyle='arc3,rad=-0.2', color='grey'))

plt.scatter(market_std_dev_2, market_return_2, marker='D', color='orange', s=100, label='Market Portfolio 2')
plt.annotate('50% Stocks, 40% Bonds, \n10% Monolith', (market_std_dev_2, market_return_2), textcoords="offset points", xytext=(-10,-10), ha='center', 
arrowprops=dict(arrowstyle='-', connectionstyle='arc3,rad=-0.2', color='grey'))


########################################################################################################################################
######################THIS SECTION HANDLES PLOT ATTRIBUTES##############################################################################
########################################################################################################################################


# plot attributes
plt.xlabel('Volatility (Risk)')
plt.ylabel('Annual Return')
plt.legend([],[], frameon=False)
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:.1%}'.format(x)))
plt.gca().xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:.1%}'.format(x)))
plt.grid(True, alpha=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.show()
```
